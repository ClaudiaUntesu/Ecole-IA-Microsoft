```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

# bivariate distribution

# La meilleure façon d'analyser la distribution bivariée
# dans seaborn est en utilisant la fonction jointplot().
# Jointplot crée une figure à plusieurs panneaux qui projette
# la relation bivariée entre deux variables et aussi
# la distribution univariée de chaque variable sur des axes distincts.

p1= sns.jointplot(x='fare',y='age',data=titanic)
p1.fig.suptitle("fare=f(age)")
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

# heatmaps

# La visualisation des données avec des cartes thermiques est un excellent moyen d'effectuer une analyse exploratoire des données, lorsque vous disposez d'un ensemble de données avec plusieurs variables.
# Les cartes thermiques peuvent révéler instantanément un modèle général dans l'ensemble de données. Et il est très facile de faire de belles cartes thermiques avec la bibliothèque Seaborn en Python.

p5= sns.heatmap(titanic.corr(), cmap = "coolwarm", annot=False, fmt=".2f").set(title='titanic.corr()')
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

# lattice/ trellis Multi-plot grids

# Une approche utile pour explorer des données de dimension moyenne consiste à dessiner plusieurs instances du même tracé sur différents sous-ensembles de votre ensemble de données.
# Cette technique est communément appelée traçage « en treillis » ou « en treillis », et elle est liée à l'idée de « petits multiples ».
# Pour utiliser ces fonctionnalités, vos données doivent être dans un Pandas DataFrame.
# La grille de facettes forme une matrice de panneaux définis par ligne et colonne en divisant les variables.
# En raison des panneaux, une seule parcelle ressemble à plusieurs parcelles. Il est très utile d'analyser toutes les combinaisons en deux variables discrètes.
# L'objet FacetGrid prend une trame de données en entrée et les noms des variables qui formeront les dimensions de ligne, de colonne ou de teinte de la grille.
# Les variables doivent être catégorielles et les données à chaque niveau de la variable seront utilisées pour une facette le long de cet axe.
# L'avantage d'utiliser Facet est que nous pouvons entrer une autre variable dans le tracé. Nous pouvons créer de nombreuses facettes de colonne et 
# les aligner avec les lignes de la grille : FacetGrid.map



p6 = sns.FacetGrid(titanic, col='sex')

p6.map(plt.hist, 'age')
p6.add_legend()
plt.show()
```

```python
#TBD

# Une analyse du nombre de survivants selon la classe dans laquelle ils ont embarqué sur le navire. 

# bar plot 

# Le barplot() montre la relation entre une variable catégorielle et une variable continue.
# Les données sont représentées dans des barres rectangulaires où la longueur de la barre représente la proportion
# de nombre de données dans cette catégorie. Bar plot représente l'estimation de la tendance centrale.

sns.barplot(x="sex", y="survived", hue="pclass", data= titanic)
plt.show()
```
