## Nettoyage des données

```python
#  supprimer les parenthèses autour des valeurs de la colonne neighborhood 
#  supprimer $  de la colonne price et afficher les valeurs en tant que float

#TBD

location_df["neighborhood"] = location_df["neighborhood"].str.replace(r"\(", "", regex=True)
location_df["neighborhood"] = location_df["neighborhood"].str.replace(r"\)", "", regex=True)
location_df["price"] = location_df["price"].str.replace(r"\$", "", regex=True)
location_df["price"] = location_df["price"].astype(float)
location_df

```

```python
# Eclater le contenu de la colonne date en colonnes: month day year et supprimer la colonne d'orig (date)
# #TBD INDEX LOCATION FOR NEW COLOMNS !!!!!!!!!!

location_df[['month','day', 'year']]= location_df.date.str.split(" ",expand=True)# rezultat este lista de 3 elemente
location_df
```

```python
location_df = location_df.drop(['date'], axis=1)# pop, del location, slicing in loc de drop
location_df
```

```python
# definir une fonction clean_bedrooms qui permet de changer la colonne bedrooms pour garder uniquement le premier entier 
# exemples: /1br - => 1 
# /3br - 1280ft² - => 3

#TBD
#location_df["bedrooms_new"] = location_df.bedrooms.str.extract('(?:(\d+)(?:br))')

def clean_bedrooms(df, column):
    df[column]= df[column].str.findall(("\d+")).str[0]
    df[column].fillna(0, inplace=True)
    df[column]= df[column].astype(int)
    df[column]= df[column].apply(lambda x :x if x < 10 else None)
    return df[column]


location_df["bedrooms"] = clean_bedrooms(location_df, "bedrooms")
location_df



```

```python
# definir une fonction clean_surface qui permet de changer la colonne sqft et garder uniquement le dernier entier
# exemple: /3br - 1280ft² - => 1280

#TBD
def clean_surface(df, column):
    df[column]= df[column].str.extract('(?:(\d+)(?:ft²))')
    df[column].fillna(0, inplace=True)
    df[column]= df[column].astype(int)
    df[column]= df[column].apply(lambda x :x if x > 100 else None)
    return df[column]


location_df["sqft"] = clean_surface(location_df, "sqft")
location_df
```
